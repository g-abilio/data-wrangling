# Pivot Table

```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv("Spotify_Youtube_Sample.csv")
df
```

```python
df.info()
```

Pivot table:

```python
df_pivot = df.pivot_table(values = ["Stream", "Licensed", "official_video"], index = ["Artist", "Album_type"], aggfunc = "sum")
df_pivot
```

Playing with the pivot table:

```python
df_pivot.loc["Pink Floyd"]
```

```python
df_pivot.loc[["Pink Floyd", "The Beatles"]]
```

Sorting the values:

```python
df_pivot.sort_values(by = "Stream", ascending = False)
```

```python
df_pivot.sort_values(by = ["Artist","Stream"], ascending = False)
```

```python
df_pivot.sort_values(by = ["Artist","Stream"], ascending = False).loc["Metallica"]
```

```python
df_pivot.sort_values(by = ["Artist","Stream"], ascending = False).loc[["Metallica", "Caetano Veloso"]]
```
