## Pivot tables:

```python
df_pivot = df.pivot_table(values = ["salary", "remote_ratio"], index = ["job_title", "company_size"], aggfunc = "mean")
```

```python
df_pivot
```

```python
df_pivot.loc["Data Scientist"]
```

```python
df_pivot.sort_values(by = "salary", ascending = False)
```

```python
df_pivot.sort_values(by = ["job_title", "salary"], ascending = False)
```

```python
df_pivot.sort_values(by = ["salary", "company_size"], ascending = False).loc[["Data Scientist", "AI Scientist"]]
```

```python
df_pivot_company_size = df.pivot_table(values = "salary", index = "company_size", aggfunc = "mean")
```

```python
df_pivot_company_size.sort_values(by = "salary", ascending = False)
```

## Data Cubes:

```python
import atoti as tt
```

```python
session = tt.Session()
```

```python
df.info()
```

```python
salary = session.read_csv("ds-salaries.csv", keys = ["job_title", "experience_level", "company_size", "company_location"])
```

```python
salary.head()
```

```python
cube = session.create_cube(salary, name = "Salary")
cube
```

```python
m = cube.measures
m
```

```python
l = cube.levels
l
```

0D - Apex

```python
cube.query(m["salary.SUM"
])
```

1D

```python
cube.query(m["salary.SUM"], levels = [l["company_size"
]])
```

```python
cube.query(m["salary.SUM"], levels = [l["company_size"
]], filter=l["company_size"] == "L")
```

2D

```python
cube.query(m["salary.SUM"], levels = [l["company_size"], l["job_title"]])
```

```python
m
```

```python
cube.query(m["salary.SUM"], levels = [l["company_size"], l["job_title"]], filter=l["job_title"] == "Data Scientist")
```

```python
cube.query(m["salary.SUM"], levels = [l["company_size"], l["job_title"]], filter=l["job_title"] == "AI Scientist")
```

```python
cube.query(m["salary.MEAN"], levels = [l["company_size"], l["job_title"]], filter=l["job_title"] == "Data Scientist")
```

```python
cube.query(m["salary.MEAN"], levels = [l["company_size"], l["job_title"]], filter=l["job_title"] == "AI Scientist")
```

3d

```python
cube.query(m["salary.MEAN"], levels = [l["company_size"], l["job_title"], l["experience_level"]])
```

```python
cube.query(m["salary.MEAN"], levels = [l["company_size"], l["job_title"], l["experience_level"]], filter = l["experience_level"] == "SE")
```
