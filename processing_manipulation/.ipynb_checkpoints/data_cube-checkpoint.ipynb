# Data Cube

```python
import atoti as tt
import pandas as pd
```

```python
session = tt.Session()
```

```python
df = pd.read_csv("Spotify_Youtube_Sample.csv")
df.head()
```

```python
df.info()
```

```python
views = session.read_csv("Spotify_Youtube_Sample.csv", keys = ["Artist", "Track", "Album"],)
views.head()
```

Creating a data cube:

```python
cube = session.create_cube(views, name = "Views")
cube
```

```python
m = cube.measures
m
```

```python
l = cube.levels
l
```

### Playing with the cube:

0d - Apex:

```python
cube.query(m["Views.SUM"])
```

1d

```python
cube.query(m["Views.SUM"], levels = [l["Artist"]])
```

```python
cube.query(m["Views.SUM"], levels = [l["Artist"]], filter=l["Artist"] == "The Beatles")
```

2d

```python
cube.query(m["Views.SUM"], levels = [l["Artist"], l["Album"]])
```

```python
cube.query(m["Views.SUM"], levels = [l["Artist"], l["Album"]], filter=l["Artist"] == "Pink Floyd")
```

```python
cube.query(m["Views.SUM"], levels = [l["Artist"], l["Track"]], filter=l["official_video"] == "true")
```

```python
cube.query(m["Views.SUM"], levels = [l["Artist"], l["Track"]], filter=l["official_video"] == "false")
```

3d

```python
cube.query(m["Views.SUM"], levels = [l["Artist"], l["Album"], l["Track"]])
```

```python
cube.query(m["Views.SUM"], levels = [l["Artist"], l["Album"], l["Track"]], filter=l["official_video"] == "false")
```

You can continue adding more dimensions, until the max dimension is reached (base).
