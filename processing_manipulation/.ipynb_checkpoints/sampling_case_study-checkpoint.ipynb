# Sampling case study

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Spotify_Youtube_Sample.csv")
df.head()
```

```python
df.info()
```

Sampling by number:

```python
df.sample(n=10)
```

Sampling by fraction:

```python
df.sample(frac=0.001)
```

Sampling with replacement:

```python
df[:10]
```

```python
df[:10].sample(frac = 0.9, replace = True)
```
