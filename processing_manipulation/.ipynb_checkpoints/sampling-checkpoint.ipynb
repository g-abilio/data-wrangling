# Sampling

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Nov2Temp.csv")
```

```python
df.head()
```

```python
df.info()
```

Sampling by number:

```python
df.sample(n = 5)
```

Sampling by percentage:

```python
df.sample(frac = 0.1)
```

```python
df.loc[:10].sample(frac = 0.5)
```

Sampling with replacement:

```python
df.loc[:10].sample(frac = 0.9, replace = True)
```

Using a bigger dataset and trying to see the distribution:

```python
df_h = pd.read_csv("par_disease copy.csv")
df_h
```

```python
df_h = df_h.sample(frac = 1, replace = True)
```

```python
df_h.reset_index(inplace = True)
df_h
```

```python
df_h["index"].hist()
```

```python
import seaborn as sns
import warnings
warnings.filterwarnings("ignore")
```

```python
sns.displot(data = df_h, x = "index", kind = "kde")
```
