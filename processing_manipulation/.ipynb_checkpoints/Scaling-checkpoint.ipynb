## Min-Max Scaling

A way of performing scaling where we are going to follow the rule below:

$ V_i = \frac{v_i - min}{max - min} \cdot (max' - min') + min' $

Where $V_i$ is the new value, $v_i$ is the old value, max the old max value, min the old min value, max' the new max value and min' the new min value. A famous way of doing this is using max' = 1 and min' = 0, that is, utilizing normalization, so the new value is on the (1,0) range. The equation will be:

$ V_i = \frac{v_i - min}{max - min}  $


```python
df["salary_min_max"] = (df["salary"] - df["salary"].min())/(df["salary"].max() - df["salary"].min())
```

```python
df["salary_min_max"].describe()
```

## Z-Score standardization/normalization

Another way to perform scaling. Here, we are going to subtract the mean from the value and divide it by the std, that is, we are going to see how far the value is from the mean using the std as a measure. This is based on the gaussian distribution:

$ V_i = \frac{v_i - mean}{std} $

The distribution of the data is:

```python
import matplotlib.pyplot as plt

plt.subplot(1, 2, 1)
plt.hist(df["salary"])
plt.xlabel("Salary")
plt.title("Salary")

plt.subplot(1, 2, 2)
plt.violinplot(df["salary"])
plt.xlabel("Salary")
plt.title("Salary")
```

```python
import seaborn as sns

sns.displot(data = df, x = "salary", kind = "kde")
```

```python
df["salary_z_score"] = (df["salary"] - df["salary"].mean())/df["salary"].std()
```

```python
df["salary_z_score"].describe()
```

```python
df["salary"].mean(), df["salary"].std(), df["salary"].min(), df["salary"].max()
```

## Decimal Scaling

In this way of scaling, our goal is to map the values into the (-1,1) range. For that, we are going to use the equation below:

$ V_i = \frac{v_i}{10 ^ t} $

Where t is the number of digits of the max absolute value of the dataset.

```python
df["salary"].max()
```

As salary is not negative, the max absolute is the .max().

```python
df["salary_decimal"] = df["salary"]/(10 ** 8)
```

```python
df["salary_decimal"].describe()
```

If we continue to perform this kind of manipulation, we are going to have an even smoothier distribution:

```python
def outliers_det_salary():
    outliers = (df["salary"][(df["salary"] > (df["salary"].mean() + 3 * df["salary"].std()))]).index
    return outliers
```

```python
def plots_salary_matplotlib(iterator):
    plt.figure(figsize = (10,10))
    plt.subplot(1, 2, 1)
    plt.hist(df["salary"])
    plt.title(iterator)
    plt.xlabel("Salary")

    plt.subplot(1, 2, 2)
    plt.violinplot(df["salary"])
    plt.xlabel("Salary")
    plt.show()
```

```python
def conditioning_dataframe():
    cond = outliers_det_salary()
    i = 1
    while len(cond) != 0:
        df.drop(cond, inplace = True)
        df["salary_z_score"] = (df["salary"] - df["salary"].mean())/df["salary"].std()
        plots_salary_matplotlib(i)  
        print(f"Iteration {i}: \n", df["salary_z_score"].describe())
        i += 1
        cond = outliers_det_salary()
```

```python
conditioning_dataframe()
```

```python
sns.displot(data = df, x = "salary_z_score", kind = "kde")
```
