# Dimension Elimination

```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Economy_of_US_na.csv")
df
```

```python
df_sample1 = df[df.columns[:2]]
df_sample1
```

```python
df_sample2 = df.drop(df.columns[2:], axis = 1)
df_sample2
```

```python
n_c = ["Year", "GDP_Growth"]
df_sample3 = df[n_c]
df_sample3
```

```python
d_n_c = "GDP_Nominal"
df_sample4 = df.drop(d_n_c, axis = 1)
df_sample4
```
