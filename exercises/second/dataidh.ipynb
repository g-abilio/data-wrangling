```python
soup = BeautifulSoup(source.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
listidh = soup.find_all("td")
lista = []
listidh
```

```python
for i in listidh:
    lista.append(i.get_text())
```

```python
newlist = []

for i in lista:
    x = i.replace(",", ".")
    newlist.append(x)
```

```python
listnumber = []

for i in newlist:
    try:
        x = float(i)
    except:
        continue
    finally:
        listnumber.append(x)
        
listnumber
```

```python
del listnumber[0]
listnumber
```

```python
finallist = []

for i in listnumber:
    if 0 < i < 1:
               finallist.append(i)
```

# Web Scraping 2021

```python
import pandas as pd
import numpy as np
import requests
```

```python
src1 = requests.get("https://countryeconomy.com/hdi")
```

```python
src1
```

```python
src1.content
```

```python
from bs4 import BeautifulSoup
```

```python
soup1 = BeautifulSoup(src1.content, "html.parser")
print(soup1.prettify())
```

```python
lista = soup1.find_all("td", class_ = "numero")
listaux = []
for i in lista:
    listaux.append(i.get_text())
    
listaux
```

```python
listidh = []

for i in range(0,len(listaux), 3):
    listidh.append(float(listaux[i]))
    
listidh
```

```python
arrayidh = np.array(listidh)
idhscraping = arrayidh.mean()
idhscraping
```
