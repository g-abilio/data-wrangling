# Combining data from different kinds of files

```python
import pandas as pd
```

```python
df1 = pd.read_csv("C2-M6-Data/boulderdaily_Part8.csv")
df1
```

```python
df1.info()
```

```python
dfs1  = df1[["year", "month", "day"]].sample(n=5)
dfs1
```

```python
dfs2  = df1[["year", "month", "day"]].sample(n=5)
dfs2
```

```python
dfs3  = df1[["year", "month", "day"]].sample(n=5)
dfs3
```

### Concatenation

```python
dfconcat = pd.concat([dfs1, dfs2, dfs3], axis = 0)
dfconcat
```

```python
dfconcat.info()
```

```python
dfs1 = df1[["year", "month", "day"]][:5]
dfs2 = df1[["tmax", "tmin"]][:5]
dfs3 = df1[["precip", "snow", "snowcover"]][:5]
```

```python
dfconcat = pd.concat([dfs1, dfs2, dfs3], axis = 1)
dfconcat
```

```python
df1.head() == dfconcat
```

```python
dfs1 = df1[["year", "tmax"]][:5]
dfs2 = df1[["tmax", "day"]][:5]
```

```python
dfs1
```

```python
dfs2
```

### Merging

```python
dfmerge = pd.merge(dfs1, dfs2, on = ["tmax"], how = "inner")
```
