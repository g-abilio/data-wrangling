# Web scraping

```python
import pandas as pd
import numpy as np
```

```python
import requests
```

```python
source = requests.get("https://www.undp.org/pt/brazil/idh-global-2014")
source
```

```python
source.status_code
```

```python
source.content
```

```python
from bs4 import BeautifulSoup
```

```python
soup = BeautifulSoup(source.content, "html.parser")
```

```python
print(soup.prettify())
```

```python
listidh = soup.find_all("td")
lista = []
listidh
```

```python
for i in listidh:
    lista.append(i.get_text())
```

```python
newlist = []

for i in lista:
    x = i.replace(",", ".")
    newlist.append(x)
```

```python
listnumber = []

for i in newlist:
    try:
        x = float(i)
    except:
        continue
    finally:
        listnumber.append(x)
        
listnumber
```

```python
listnumber
```

```python
finallist = []

for i in listnumber:
    if 0 < i < 1:
               finallist.append(i)
```

```python
finallist
```

```python
finallist.index(0.348)
```

```python
finallist[190]
```

```python
while len(finallist) > 191:
    del finallist[190]
```

```python
finallist
```

```python
array = np.array(finallist)
type(array)
```

```python
(array.mean(), array.std(), array.max(), array.min())
```
