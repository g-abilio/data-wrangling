# Using Yahoo Finance API

```python
import yfinance as yahooFinance
```

```python
googleinfo = yahooFinance.Ticker("GOOGL")
```

```python
print(googleinfo.info)
```

```python
infod = googleinfo.info
infod["zip"]
```

```python
for k, v in infod.items():
    print(k, ":", v)
```

```python
display(googleinfo.history(period = "max"))
```

```python
import datetime

start = datetime.datetime(2004, 10, 21)
end = datetime.datetime(2017, 9, 2)
display(googleinfo.history(start = start, end = end))
```

```python
df = yahooFinance.download("AMZN GOOGL", start = start, end = end, group_by = "ticker")
df
```

```python
df.to_csv("csv_google_amazon.csv")
```
