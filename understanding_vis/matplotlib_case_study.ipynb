## Plotting both attributes simultaneously

```python
plt.plot(df_2000_1month["day"], df_2000_1month["tmax"], label = "Tmax")
plt.plot(df_2000_1month["day"], df_2000_1month["tmin"], label = "Tmin")
plt.xlabel("Day")
plt.title("Climate January of 2000")
plt.legend()
```

```python
plt.subplot(2,1,1)
plt.plot(df_2000_1month["day"], df_2000_1month["tmax"], ":r")
plt.ylabel("Tmax")
plt.title("Climate January of 2000")
plt.grid()

plt.subplot(2,1,2)
plt.plot(df_2000_1month["day"], df_2000_1month["tmin"], ":b")
plt.xlabel("Day")
plt.ylabel("Tmin")
plt.grid()
```

```python
plt.scatter(df_2000_1month["day"], df_2000_1month["tmax"], c = df_2000_1month["tmin"], cmap = "hot")
plt.colorbar()
plt.ylabel("Tmax")
plt.xlabel("Day")
plt.title("Climate January of 2000")
```

```python
plt.scatter(df_2000_1month["day"], df_2000_1month["tmax"], alpha = 0.5, s = df_2000_1month["tmin"]*5)
plt.ylabel("Tmax")
plt.xlabel("Day")
plt.title("Climate January of 2000")
```

Plotting the thermal amplitude:

```python
plt.bar(df_2000_1month["day"], df_2000_1month["tmax"] - df_2000_1month["tmin"])
plt.xlabel("Day")
plt.ylabel("Thermal amplitude")
plt.title("Climate January of 2000")
```

```python
plt.scatter(df_2000_1month["day"], df_2000_1month["tmax"], c = df_2000_1month["tmin"], cmap = "hot", alpha = 0.5, s = (df_2000_1month["tmax"] - df_2000_1month["tmin"]) * 4)
plt.colorbar()
plt.xlabel("Day")
plt.ylabel("Tmax")
plt.title("Climate January of 2000")
```

Getting the analysis more into the Thermal Amplitude:

```python
plt.hist(df_2000_1month["tmax"] - df_2000_1month["tmin"])
plt.xlabel("Thermal Amplitude")
plt.ylabel("Count")
plt.title("Climate January of 2000")
```

Increasing the range of the analysis by changing the size of the dataframe in focus, we can see that the histogram is fitting the Gaussian curve better every time:

```python
for i in range(500, len(df), 500):
    plt.hist(df[0:i]["tmax"] - df[0:i]["tmin"])
    plt.xlabel("Thermal Amplitude")
    plt.ylabel("Count")
    plt.title("Climate")
    plt.show()
```

```python
for i in range(500, len(df), 500):
    plt.violinplot(df[0:i]["tmax"] - df[0:i]["tmin"])
    plt.ylabel("Thermal Amplitude")
    plt.title("Climate")
    plt.show()
```
