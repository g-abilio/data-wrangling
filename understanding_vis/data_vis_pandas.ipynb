### Box plot

Here, the lower part of the box represents the first quartile (Q1, p = 0.25), the green line at the center of the box represents the second quartile (Q2, p = 0.5), and the upper bound represents the third quartile (Q3, p = 0.75). The arrow going up from the upper bound is calculate by: Q3 + 1.5 * IQR (Inter Quantile Range = Q3 - Q1). The lower arrow is calculated by: Q1 - 1.5 * IQR. The data inside the range determined by those arrows is considered normal, as the upper arrow marks the theoretical maximum value, and the lower arrow marks the theoretical minimum value. The data outside this region is considered abnormal.

```python
df["Inflation"].plot(kind = "box")
```

```python
df["Unemployment"].plot(kind = "box") 
```
