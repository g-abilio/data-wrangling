## Plotting

We are going to skip the scattering and the color map, because it doesn't make sense using these tools here, as the results of the plottings are very bad.

```python
non_numerical_values = set(df.columns) - {"salary", "salary_in_usd", "remote_ratio"}
non_numerical_values
```

```python
numerical_values = set(df.columns) - non_numerical_values
numerical_values
```

### Line

```python
for i in numerical_values:
        df.plot(y = i)
```

### Area Plotting

```python
for i in numerical_values:
    df.plot(y = i, kind = "area")
```

### Bar charts

Not displays a great result in respect of the distribution of the numerical values on the dataframe according to the index as the number f indexes is to high.

```python
for i in numerical_values:
    df.plot(y = i, kind = "bar")
```
