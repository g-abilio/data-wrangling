# Uploading the CSV

```python
import pandas as pd
import seaborn as sns
```

```python
us_economy = pd.read_csv("Economy_of_US.csv")
```

```python
us_economy
```

```python
us_economy.info()
```

Here, we have just one column as object. To get a better grasp of the dataset in the plots, we are going to add another column:

```python
list_aux = []

for x in us_economy["GDP_Growth"]:
    if x > 0: 
        list_aux.append("Increase")
    elif x < 0:
        list_aux.append("Decrease")
    else:
        list_aux.append("No change")
        
list_aux
```

```python
us_economy["GDP_Change"] = list_aux
```

```python
us_economy
```

```python
us_economy.info()
```

```python
sns.relplot(data = us_economy, x = "Year", y = "Inflation", hue = "Inflation_Change", col = "GDP_Change")
```

```python
sns.relplot(data = us_economy, x = "Year", y = "Inflation", hue = "Inflation_Change", col = "GDP_Change", size = "Unemployment", sizes = (10, 200), alpha = (0.6))
```

```python
sns.relplot(data = us_economy, x = "Inflation_Change", y = "GDP_Growth")
```

In the next plot, the shadow region represents 95% of the possible values (the mean value +- 2 times the std).

```python
sns.relplot(data = us_economy, x = "Inflation_Change", y = "GDP_Growth", kind = "line")
```
