# Opening the CSV

```python
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("Economy_of_US.csv")
```

```python
df
```

```python
df.describe()
```

```python
df.info()
```

## Simple plots

```python
plt.plot(df["Year"], df["GDP_Growth"])
plt.xlabel("Year")
plt.ylabel("GDP Growth")
plt.title("Economy of US")
```

### Using markers

```python
plt.plot(df["Year"], df["GDP_Growth"], "o")
plt.xlabel("Year")
plt.ylabel("GDP Growth")
plt.title("Economy of US")
```

```python
from matplotlib.lines import Line2D
markers = list(Line2D.markers.keys())
```

```python
markers
```

```python
for i in markers:
    plt.plot(df["Year"], df["GDP_Growth"], i)
    plt.xlabel("Year")
    plt.ylabel("GDP Growth")
    plt.title("Economy of US")
    plt.show()
```

### Line styles

```python
linestyles = list(Line2D.lineStyles.keys())
```

```python
del linestyles[4]
linestyles
```

```python
for i in linestyles:
    l = "o" + i + "r"
    plt.plot(df["Year"], df["GDP_Growth"], l)
    plt.xlabel("Year")
    plt.ylabel("GDP Growth")
    plt.title("Economy of US")
    plt.show()
```

### Colors

```python
from matplotlib import colors
```

```python
colors = list(colors.BASE_COLORS.keys())
```

```python
colors
```

```python
for i in colors:
    l = "o" + ":" + i
    plt.plot(df["Year"], df["GDP_Growth"], l)
    plt.xlabel("Year")
    plt.ylabel("GDP Growth")
    plt.title("Economy of US")
    plt.show()
```

### Marker size

```python
plt.plot(df["Year"], df["GDP_Growth"], "o--", ms = 10)
plt.xlabel("Year")
plt.ylabel("GDP Growth")
plt.title("Economy of US")
```

## Scatter Plot

```python
plt.scatter(df["Year"], df["GDP_Growth"])
plt.xlabel("Year")
plt.ylabel("GDP_Growth")
plt.title("Economy of US")
```

### Colorbar

```python
plt.scatter(df["Year"], df["GDP_Growth"], c = df["Inflation"], cmap = "viridis")
plt.colorbar()
plt.xlabel("Year")
plt.ylabel("GDP_Growth")
plt.title("Economy of US")
```

### Size

```python
plt.scatter(df["Year"], df["GDP_Growth"], s = df["Inflation"]*1000)
plt.xlabel("Year")
plt.ylabel("GDP Growth")
plt.title("Economy of US")
```

 ### Colorbar and size

```python
plt.scatter(df["Year"], df["GDP_Growth"], c = df["Inflation"], alpha = 0.5, cmap = "viridis", s = df["Unemployment"]*1000)
plt.colorbar()
plt.xlabel("Year")
plt.ylabel("GDP Growth")
plt.title("Economy of US")
```

## Bar plot

```python
plt.bar(df["Year"], df["GDP_Growth"])
plt.xlabel("Year")
plt.ylabel("GDP Growth")
plt.title("Economy of US")
```

## Histogram

```python
plt.hist(df["Inflation"])
plt.xlabel("Inflation")
plt.ylabel("Count")
plt.title("Economy of US")
```

```python
plt.hist(df["Inflation"], bins = 50)
plt.xlabel("Inflation")
plt.ylabel("Count")
plt.title("Economy of US")
```

## Pie Plot

```python
df["Inflation_Change"].value_counts()
```

```python
plt.pie(df["Inflation_Change"].value_counts(), labels = ["Decrease", "Increase", "No change"])
plt.legend()
plt.title("Inflation Change")
```

```python
plt.pie(df["Inflation_Change"].value_counts(), labels = ["Decrease", "Increase", "No change"], explode = [0, 0, 0.1])
plt.legend()
plt.title("Inflation Change")
```

## Box plot

```python
plt.boxplot(df["Inflation"])
plt.ylabel("Inflation")
plt.title("Economy of US")
```

## Violin plot

Similar to the kde one on pandas.

```python
plt.violinplot(df["Inflation"])
plt.ylabel("Inflation")
plt.title("Economy of US")
```

## Multiplots

```python
df.info()
```

```python
plt.plot(df["Year"], df["GDP_Growth"])
plt.plot(df["Year"], df["Inflation"])
plt.plot(df["Year"], df["Unemployment"])
plt.xlabel("Year")
plt.title("Economy of US")
```

```python
plt.plot(df["Year"], df["GDP_Growth"], label = "GDP Growth")
plt.plot(df["Year"], df["Inflation"], label = "Inflation")
plt.plot(df["Year"], df["Unemployment"], label = "Unemployment")
plt.xlabel("Year")
plt.ylabel("Economy")
plt.legend()
plt.grid()
plt.title("Economy of US")
```

## Different Arrangement

```python
plt.subplot(3,1,1)
plt.plot(df["Year"], df["GDP_Growth"], label = "GDP Growth")
plt.xlabel("Year")
plt.ylabel("GDP Growth")
plt.legend()
plt.grid()

plt.subplot(3,1,2)
plt.plot(df["Year"], df["Inflation"], label = "Inflation")
plt.xlabel("Year")
plt.ylabel("Inflation")
plt.legend()
plt.grid()

plt.subplot(3,1,3)
plt.plot(df["Year"], df["Unemployment"], label = "Unemployment")
plt.xlabel("Year")
plt.ylabel("Unemployment")
plt.legend()
plt.grid()
```

```python
plt.subplot(1,3,1)
plt.plot(df["Year"], df["GDP_Growth"], label = "GDP Growth")
plt.xlabel("Year")
plt.ylabel("Economy")
plt.legend()
plt.grid()

plt.subplot(1,3,2)
plt.plot(df["Year"], df["Inflation"], label = "Inflation")
plt.xlabel("Year")
plt.legend()
plt.grid()

plt.subplot(1,3,3)
plt.plot(df["Year"], df["Unemployment"], label = "Unemployment")
plt.xlabel("Year")
plt.legend()
plt.grid()
```
